```python
import cueBeamCore
import numpy
import matplotlib.pyplot as plt
q=cueBeamCore.CueBeamSolver()
q.elements
q.wavenumber = 1/(1e-3)
q.rxPlane.x0= 0.01e-3
q.rxPlane.y0 = 1.0e-3
q.rxPlane.z0 = 0.0e-3
q.rxPlane.dy = 0.5e-3
q.rxPlane.dz = 0.5e-3
q.rxPlane.ny = 100
q.rxPlane.nz = 100
q.rxPlane.clear_pressurefield()
q.beamsim()
q.rxPlane.pressurefield
imgplot = plt.imshow(
    X=numpy.real(q.rxPlane.pressurefield),
    extent=(q.rxPlane.z0,q.rxPlane.nz*q.rxPlane.dz,q.rxPlane.y0,q.rxPlane.y0+q.rxPlane.ny*q.rxPlane.dy),
    interpolation = "spline36")
plt.show()
```
